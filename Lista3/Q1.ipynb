```python
# importando a biblioteca numpy para fazer a lista de números
import numpy as np 
# importando a biblioteca matplotlib.pyplot para fazer o grafico
import matplotlib.pyplot as plt 
# forma de exibicao do grafico
%matplotlib inline 
# importando a biblioteca math para poder utilizar a funcao math.pi
import math

x = np.arange(0,72.5,0.5) # x de 0 72 
y = 24 + 3 * np.cos((x*math.pi/6) + math.pi/3) # funcao y
plt.plot(x,y) # desenho da funcao
plt.grid() # grade
plt.xlabel("tempo (horas)") # legenda do eixo x
plt.ylabel("temperatura (°C)") # legenda do eixo y

```
