```python
# importando a biblioteca numpy para fazer a lista de números
import numpy as np 
# importando a biblioteca matplotlib.pyplot para fazer o grafico
import matplotlib.pyplot as plt 
# forma de exibicao do grafico
%matplotlib inline 
# importando a biblioteca math para poder utilizar a funcao math.pi
import math

x = np.arange(-math.pi/4, math.pi/2, 0.1) # x tem inicio em -π/4  e fim em π/2

# funcoes dos itens a ao e
a = np.sin(x)
b = np.cos(2*x)
c = 2*np.sin(x)
d = np.sin(2*x)
e = np.cos(x)

# desenhos dos graficos a ao e
plt.plot(x, a, 'y', label='a')
plt.plot(x, b, 'purple', label='b')
plt.plot(x, c, 'orange', label='c')
plt.plot(x, d, 'g', label='d')
plt.plot(x, e, 'r', label='e')

plt.grid() # grade
plt.xlabel("eixo x") # legenda do eixo x
plt.ylabel("eixo y") # legenda do eixo y
plt.legend() # legenda
```

```python
# para melhor comparacao

x = np.arange(-math.pi/4, math.pi/2, 0.1) # x tem inicio em -π/4  e fim em π/2

# funcoes dos itens b e d
b = np.cos(2*x)
d = np.sin(2*x)

# desenho do grafico b e d
plt.plot(x, b, 'purple', label='b')
plt.plot(x, d, 'g', label='d')

plt.grid() # grade
plt.xlabel("eixo x") # legenda do eixo x
plt.ylabel("eixo y") # legenda do eixo y
plt.legend() # legenda
```
