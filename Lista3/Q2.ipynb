```python
# importando a biblioteca numpy para fazer a lista de números
import numpy as np 
# importando a biblioteca matplotlib.pyplot para fazer o grafico
import matplotlib.pyplot as plt 
# forma de exibicao do grafico
%matplotlib inline 
# importando a biblioteca math para poder utilizar a funcao math.pi
import math

x = np.arange(-2*math.pi, 2*math.pi + 0.1, 0.1) # x tem inicio em -2π  e fim em 2π 
y1 = np.sin(x) # funcao y1
y2 = np.cos(x) # funcao y2

plt.plot(x, y1, 'r') # desenho do grafico y1 em vermelho
plt.plot(x, y2, 'y') # desenho do grafico y2 em amarelo

plt.grid() # grade
plt.xlabel("eixo x") # legenda do eixo x
plt.ylabel("eixo y") # legenda do eixo y
```
