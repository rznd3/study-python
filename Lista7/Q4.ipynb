```python
import pandas as pd

df = pd.read_csv("https://www.dropbox.com/s/kwibw8umms9yh9w/NotasAjuste.csv?dl=1")

exame = input("")

med = df[exame].mean() 

if med < 50:
  df["Ajuste"] = df[exame] + df[exame]/5
elif med >= 50 and med <= 70:
   df["Ajuste"] = df[exame] + df[exame]/10
elif med >= 70 and med <= 80:
   df["Ajuste"] = df[exame] + df[exame]/20
else:
  df["Ajuste"] = df[exame]
  
print(df)
```
