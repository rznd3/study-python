```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://www.dropbox.com/s/g3d4mcd9f8gqkjh/fake-classrooms-correl17.csv?dl=1")

col1 = input("")
col2 = input("")

# correlacao
correlacao = df[col1].corr(df[col2])
print("Correlao entre", col1, "e", col2, " = ", "%.2f" %correlacao)

# equacao
print("y =", col1, "x", "+", col2)

(a, b) = np.polyfit(x = df[col1], y = df[col2], deg = 1)
print("Equao:", "y = ", "%.2fx + %.2f" % (a, b))

# parte d predicao
x = float(input())
y = a*x + b

print("Predio de", col1, "para", col2, " = ", "%.2f" %x, "", "%.2f" %y, ".")

if abs(correlacao) < 0.38:
  print("Correlao < 0.38. Predio no exibida!")
elif abs(correlacao) >= 0.83:
  print("Correlao >= 0.83.")
elif abs(correlacao) >= 0.38 and abs(correlacao) < 0.83:
  print("Correlao < 0.83. Ateno com a predio")
```
