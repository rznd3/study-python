```python
import pandas as pd

df = pd.read_csv("https://www.dropbox.com/s/uno2gzc5618fmxi/fake-classrooms07.csv?dl=1")

indice_aluno = int(input(""))

nome, prova1, prova2, trabalho = df[["Nome","Prova 1","Prova 2","Trabalho"]].loc[indice_aluno]
print("Nome:", nome)
print("Prova 1: ", prova1)
print("Prova 2: ", prova2)
print("Trabalho: ", trabalho)

med = (prova1 + prova2 + trabalho)/3 
print("Media = %.2f"%med)

if med >= 5.7:
  print("Aprovado")
else:
  print("Reprovado")
```
