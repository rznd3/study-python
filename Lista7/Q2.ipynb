```python
import pandas as pd

df = pd.read_csv("https://www.dropbox.com/s/qgp1gqxr468qj7u/fake-classrooms10.csv?dl=1")

coluna_lida1 = input("")
coluna_lida2 = input("")

media1 = df[coluna_lida1].mean()
media2 = df[coluna_lida2].mean()

print("Media", coluna_lida1, " = ", "%.2f" %media1)
print("Media", coluna_lida2, " = ", "%.2f" %media2)

if media1 > 7 and media2 < 7:
  print("Apenas Media", coluna_lida1, "eh maior que 7.0")

if media2 > 7 and media1 < 7:
  print("Apenas Media", coluna_lida2, "eh maior que 7.0")

if media1 and media2 > 7:
  print("As duas Medias sao maiores que 7.0")
else:
  print("Nenhuma das medias eh maior que 7.0")
```
