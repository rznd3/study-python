```python
import pandas as pd
df = pd.read_csv("https://www.dropbox.com/s/6qwgc2dgyat96qd/fertilidade.csv?dl=1", sep=";")

lido = input()

df_filtrada = df.query("Regio == '"+lido+"' and Contraceptivos <= 47")

media = df_filtrada["Fertilidade"].mean()     
mediana = df_filtrada["Fertilidade"].median()     
variancia = df_filtrada["Fertilidade"].var()     

print("Fertilidade na", lido, "com Contraceptivos <= 47:")                                  
print("media: %.2f," %media, end=" ")
print("mediana: %.2f," %mediana, end=" ")
print("variancia: %.2f," %variancia)


```
