```python
import pandas as pd 
import numpy as np

df = pd.read_csv("https://www.dropbox.com/s/7k249harn2sqlgq/Exames3_versao4.csv?dl=1") 

(a, b, c, d) = np.polyfit(x=df["Exame 1"], y=df["Exame 2"], deg = 3) 

n = int(input("Digite um valor inteiro: "))

predicao = a*n**3+b*n**2+c*n+d
print("%.2f" %predicao)
```
