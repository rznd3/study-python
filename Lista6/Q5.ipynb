```python
import pandas as pd  
df = pd.read_csv("https://www.dropbox.com/s/j6iinykmyckhv74/fake-classrooms-correl19.csv?dl=1") 

coluna = input("Digite a coluna para calcular o coeficiente de determinação (Faltas ou Horas Estudo): ")

cc  = df['Nota Final'].corr(df[coluna]) 
cd = cc**2

print("%.2f" %cd)
```
