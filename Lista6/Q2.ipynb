```python
import pandas as pd 
import numpy as np

df = pd.read_csv("https://www.dropbox.com/s/ahlznbnx07dfpqt/Exames3.csv?dl=1") 

# calculamos os valores de a e b para a reta ax+b de regressão linear com Trabalho no eixo x e Prova 2 no eixo y
(a, b) = np.polyfit(x=df["Exame 2"], y=df["Exame 3"], deg = 1) 

n = int(input("Digite um valor inteiro: "))

predicao = a*n+b
print("%.2f" %predicao)
```
